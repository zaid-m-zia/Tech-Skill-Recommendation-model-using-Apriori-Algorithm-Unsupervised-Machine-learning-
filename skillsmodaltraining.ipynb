```python
df['LanguageWorkedWithCleaned'] = df['LanguageHaveWorkedWith'].fillna('').str.split(';').apply(lambda x: [skill.strip().lower() for skill in x] if x is not None else []).apply(lambda x: [skill for skill in x if skill])
```

### Explanation of the code:

1.  **`df['LanguageWorkedWithCleaned'] = ...`**: We are creating a new column called `LanguageWorkedWithCleaned` to store the cleaned data, so we don't modify the original column.
2.  **`df['LanguageWorkedWith'].str.split(';')`**: This part takes the `LanguageWorkedWith` column and splits each entry (which is a string of skills separated by semicolons) into a list of strings. For example, 'Python;JavaScript' becomes `['Python', 'JavaScript']`.
3.  **`.apply(lambda x: [skill.strip().lower() for skill in x] if x is not None else [])`**: This applies a function to each of these lists:
    *   **`if x is not None else []`**: Checks if the entry is not empty (not `None`). If it's `None` (meaning no languages were worked with), it creates an empty list.
    *   **`[skill.strip().lower() for skill in x]`**: If the entry is not `None`, it iterates through each skill in the list. For each `skill`, it `strip()`s any leading or trailing whitespace (like spaces) and then converts it to `lower()`case. This ensures consistency (e.g., 'Python' and 'python' are treated as the same skill).
4.  **`.apply(lambda x: [skill for skill in x if skill])`**: This final `apply` step filters out any empty strings that might have resulted from the splitting and cleaning process (e.g., if there were extra semicolons like 'Python;;JavaScript', splitting would result in `['Python', '', 'JavaScript']`, and this step removes the empty string).

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)

import pandas as pd
df_encoded = pd.DataFrame(te_array, columns=te.columns_)

df_encoded.head()
```

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets = apriori(df_encoded, min_support=0.02, use_colnames=True)

frequent_itemsets.head()
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)

rules.head()
```

```python
def recommend_skills(user_skills, rules):
    user_skills = set([s.lower() for s in user_skills])

    recommendations = set()

    for _, row in rules.iterrows():
        antecedents = set(row['antecedents'])
        consequents = set(row['consequents'])

        if antecedents.issubset(user_skills):
            recommendations.update(consequents)

    return list(recommendations)
```

```python
recommend_skills(["assembly"], rules)
```

```python
def recommend_skills(user_skills, rules):
    user_skills = set([s.lower() for s in user_skills])

    recommendations = set()

    for _, row in rules.iterrows():
        antecedents = set(row['antecedents'])
        consequents = set(row['consequents'])

        if antecedents.issubset(user_skills):
            recommendations.update(consequents)

    # remove already known skills
    recommendations = recommendations - user_skills

    return list(recommendations)
```

```python
rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=0.4  # increase from 0.3
)

# Also filter by lift
rules = rules[rules['lift'] > 1.2]
```

```python
print(len(rules))
```

```python
recommend_skills(["python"], rules)
```

```python
result = recommend_skills(["java"], rules)
print(result)
```
